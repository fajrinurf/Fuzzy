## Control System of Window Automation with Fuzzy Logic


--------------------------------------
Control System ini bertujuan agar cahaya matahari dapat secara otomatis masuk ke dalam rumah / kantor yang 
dalam beberapa kasus akan membantu penghuni bangun atau pun dapat membantu mendapatkan sinar matahari dan juga
berfungsi sebagai sistem keamanan yaitu agar terlihat bahwa sedang ada aktivitas didalam rumah / kantor jika
sedang ditinggalkan dalam waktu lama.

### Input
Waktu
- **Subuh** : 0 - 5
- **Pagi**  : 6 - 10
- **Siang** : 11 - 14
- **Sore**  : 15 - 17
- **Malam** : 18 - 24
    
Kelembaban Ruangan
- **Kering** : 0% - 30%
- **Lembab** : 40% - 60%
- **Basah**  : 70% - 100%
    
Suhu Luar Ruangan
- **Dingin** : (-10) - 14
- **Normal** : 15 - 27
- **Hangat** : 28 - 41
- **Panas**  : 42 - 60
    
### Output

Tirai Jendela Terbuka

Tirai Jendela Tertutup

# METHOD WAKTU [↑](#up) <a class="anchor" id="Method_Waktu"></a> 

## PARAMETER WAKTU

```python
def setparameterwaktu():
    print("Subuh < Subuh - Pagi < Pagi - Siang < Siang < Siang - Sore < Sore < Sore - Malam < Malam")
    sbh = 5
    sbpg = 7
    pgsg = 10
    sg = 11
    sgsr = 14
    sr = 15
    srmm = 17
    mlm = 18
    
    return sbh,sbpg,pgsg,sg,sgsr,sr,srmm,mlm
```

## Fungsi Keanggotaan Waktu [↑](#up) <a class="anchor" id="Anggota_Waktu"></a>

### Subuh [↑](#up) <a class="anchor" id="Subuh"></a>

```python
def wktsbh(waktuin) :
    nilai_wktsbh = 0
    
    if (waktuin < 5 ): 
        nilai_wktsbh = 1
    elif (waktuin < 7): 
        nilai_wktsbh = (-((waktuin - 7) / (7 - 5)))
    else: 
        nilai_wktsbh = 0
    
    return nilai_wktsbh
```

```python
def waktusubuh(x,sbh,sbpg):
    if (x < sbh ): 
      return 1
    elif (x < sbpg): 
      return (-((x - sbpg) / (sbpg - sbh)))
    else: 
      return 0
```

### Pagi [↑](#up) <a class="anchor" id="Pagi"></a>

```python
def wktpg(waktuin):
    nilai_wktpg = 0
    
    if (5 < waktuin <= 7): 
        nilai_wktpg = ((waktuin - 5) / (7 - 5))
    elif (7 < waktuin <= 10):
        nilai_wktpg = 1
    elif (10 < waktuin <= 11): 
        nilai_wktpg = (-((waktuin - 11) / (11 - 10)))
    else: 
        nilai_wktpg = 0
        
    return nilai_wktpg
```

```python
def waktupagi(x,sbh,sbpg,pgsg,sg):
    if (sbh < x <= sbpg): 
      return ((x - sbh) / (sbpg - sbh))
    elif (sbpg < x <= pgsg):
      return 1
    elif (pgsg < x <= sg): 
      return (-((x - sg) / (sg - pgsg)))
    else: 
      return 0
```

### Siang [↑](#up) <a class="anchor" id="Siang"></a>

```python
def wktsg(waktuin):
    nilai_wktsg = 0
    
    if (10 < waktuin <= 11): 
        nilai_wktsg = ((waktuin - 10) / (11 - 10))
    elif (11 < waktuin <= 14):
        nilai_wktsg = 1
    elif (14 < waktuin <= 15): 
        nilai_wktsg = (-((waktuin - 15) / (15 - 14)))
    else: 
        nilai_wktsg = 0
        
    return nilai_wktsg
```

```python
def waktusiang(x,pgsg,sg,sgsr,sr):
    if (pgsg < x <= sg): 
      return ((x - pgsg) / (sg - pgsg))
    elif (sg < x <= sgsr):
      return 1
    elif (sgsr < x <= sr): 
      return (-((x - sr) / (sr - sgsr)))
    else: 
      return 0
```

### Sore [↑](#up) <a class="anchor" id="Sore"></a>

```python
def wktsr(waktuin):
    nilai_wktsr = 0
    
    if (14 < waktuin <= 15): 
        nilai_wktsr =  ((waktuin - 14) / (15 - 14))
    elif (15 < waktuin <= 17):
        nilai_wktsr =  1
    elif (17 < waktuin <= 18): 
        nilai_wktsr =  (-((waktuin - 18) / (18 - 17)))
    else: 
        nilai_wktsr =  0
        
    return nilai_wktsr
```

```python
def waktusore(x,sgsr,sr,srmm,mlm):
    if (sgsr < x <= sr): 
      return ((x - sgsr) / (sr - sgsr))
    elif (sgsr < x <= srmm):
      return 1
    elif (srmm < x <= mlm): 
      return (-((x - mlm) / (mlm - srmm)))
    else: 
      return 0
```

### Malam [↑](#up) <a class="anchor" id="Malam"></a>

```python
def wktmlm(waktuin):
    nilai_wktmlm = 0
    
    if (waktuin < 17): 
        nilai_wktmlm = 0
    elif (waktuin > 18): 
        nilai_wktmlm = 1
    else: 
        nilai_wktmlm = ((waktuin - 17) / (18 - 17))
        
    return nilai_wktmlm
```

```python
def waktumalam(x,srmm,mlm):
    if (x < srmm): 
      return 0
    elif (x > mlm): 
      return 1
    else: 
      return ((x - srmm) / (mlm - srmm))
```

## Fungsi Grafik Waktu [↑](#up) <a class="anchor" id="Grafik_Waktu"></a>

```python
def nilai_grafikwaktu(waktuin):

    print("Nilai Fuzzy Waktu")

    if wktsbh(waktuin) != 0:
        print("Subuh = " + str(wktsbh(waktuin)))
    if wktpg(waktuin) != 0:
        print("Pagi = " + str(wktpg(waktuin)))
    if wktsg(waktuin) != 0:
        print("Siang = " + str(wktsg(waktuin)))
    if wktsr(waktuin) != 0:
        print("Sore = " + str(wktsr(waktuin)))
    if wktmlm(waktuin) != 0:
        print("Malam = " + str(wktmlm(waktuin)))
```

## Fungsi Keanggotaan Kelembaban [↑](#up) <a class="anchor" id="Anggota_Kelembaban"></a>

### Kering [↑](#up) <a class="anchor" id="Kering"></a>

```python
def angkrg(kelembabanin):
    nilai_kering = 0
    
    if (kelembabanin < 35 ): 
        nilai_kering = 1
    elif (kelembabanin < 45): 
        nilai_kering = (-((kelembabanin - 45) / (45 - 35)))
    else: 
        nilai_kering = 0

    return nilai_kering
```

```python
def kering(x,krg,krlb):
    if (x < krg ): 
      return 1
    elif (x < krlb): 
      return (-((x - krlb) / (krlb - krg)))
    else: 
      return 0
```

### Normal [↑](#up) <a class="anchor" id="Lembab"></a>

```python
def angnorm(kelembabanin):
    nilai_lembab = 0
    
    if (35 < kelembabanin <= 45): 
        nilai_lembab = ((kelembabanin - 35) / (45 - 35))
    elif (45 < kelembabanin <= 50):
        nilai_lembab = 1
    elif (50 < kelembabanin <= 65): 
        nilai_lembab = (-((kelembabanin - 65) / (65 - 50)))
    else: 
        nilai_lembab = 0

    return nilai_lembab
```

```python
def normal(x,krg,krlb,lbbh,bsh):
    if (krg < x <= krlb): 
      return ((x - krg) / (krlb - krg))
    elif (krlb < x <= lbbh):
      return 1
    elif (lbbh < x <= bsh): 
      return (-((x - bsh) / (bsh - lbbh)))
    else: 
      return 0
```

### Lembab [↑](#up) <a class="anchor" id="Basah"></a>

```python
def anglmb(kelembabanin):
    nilai_basah = 0
    
    if (kelembabanin < 50): 
        nilai_basah = 0
    elif (kelembabanin > 65): 
        nilai_basah = 1
    else: 
        nilai_basah = ((kelembabanin - 50) / (65 - 50))

    return nilai_basah
```

```python
def lembab(x,lbbh,bsh):
    if (x < lbbh): 
      return 0
    elif (x > bsh): 
      return 1
    else: 
      return ((x - lbbh) / (bsh - lbbh))
```

### Normal [↑](#up) <a class="anchor" id="Normal"></a>

# METHOD CAHAYA MATAHARI [↑](#up) <a class="anchor" id="Method_Matahari"></a>

## PARAMETER KELEMBABAN

```python
def setparametercahaya():
    print("Gelap < Gelap - Redup < Redup - Cerah < Cerah")
    glp = 400
    glrp = 1000
    rpcr = 10000
    crh = 25000
    
    return glp,glrp,rpcr,crh
```

```python
def anggel(cahayain):
    nilai_gelap = 0
    
    if (cahayain < 400 ): 
        nilai_gelap = 1
    elif (cahayain < 1000): 
        nilai_gelap = (-((cahayain - 1000) / (1000 - 400)))
    else: 
        nilai_gelap = 0

    return nilai_gelap
```

```python
def angrdp(cahayain):
    nilai_redup = 0

    if (cahayain<=400) or (cahayain>=1000):
        nilai_redup=0
    elif (400 < cahayain < 1000):
        nilai_redup=(cahayain-400)/(1000-400)
    elif(1000 < cahayain < 10000):
        nilai_redup=(-cahayain+10000)/(10000-1000)
    else:
        nilai_redup=1

    return nilai_redup
```

```python
def kdctutup(sampel):
    if (sampel < 50 ): 
        return nilai_tutup
    elif (sampel < 80): 
        return (-((sampel - 80) / (80 - 50)))
    else: 
        return nilai_buka
```

```python
def kdstutup(sampel):
    if (sampel < 50 ): 
      return 1
    elif (sampel < 80): 
      return (-((sampel - 80) / (80 - 50)))
    else: 
      return 0
```

```python
def kdtutup(x,ttp,bkk):
    if (x < ttp ): 
      return 1
    elif (x < bkk): 
      return (-((x - bkk) / (bkk - ttp)))
    else: 
      return 0
```

```python
def kdsbuka(sampel):
    if (sampel < 50 ): 
      return 0
    elif (sampel > 80): 
      return 1
    else: 
      return (((sampel - 50) / (80 - 50)))
```

```python
def kdbuka(x,ttp,bkk):
    if (x < ttp ): 
      return 0
    elif (x > bkk): 
      return 1
    else: 
      return (-((x - ttp) / (ttp - bkk)))
```
