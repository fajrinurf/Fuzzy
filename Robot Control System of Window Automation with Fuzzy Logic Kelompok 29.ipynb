## Control System of Window Automation with Fuzzy Logic


--------------------------------------
Control System ini bertujuan agar cahaya matahari dapat secara otomatis masuk ke dalam rumah / kantor yang 
dalam beberapa kasus akan membantu penghuni bangun atau pun dapat membantu mendapatkan sinar matahari dan juga
berfungsi sebagai sistem keamanan yaitu agar terlihat bahwa sedang ada aktivitas didalam rumah / kantor jika
sedang ditinggalkan dalam waktu lama.

### Input
Waktu
- **Subuh** : 0 - 5
- **Pagi**  : 6 - 10
- **Siang** : 11 - 14
- **Sore**  : 15 - 17
- **Malam** : 18 - 24
    
Kelembaban Ruangan
- **Kering** : 0% - 30%
- **Lembab** : 40% - 60%
- **Basah**  : 70% - 100%
    
Suhu Luar Ruangan
- **Dingin** : (-10) - 14
- **Normal** : 15 - 27
- **Hangat** : 28 - 41
- **Panas**  : 42 - 60
    
### Output

Tirai Jendela Terbuka

Tirai Jendela Tertutup

# METHOD WAKTU [↑](#up) <a class="anchor" id="Method_Waktu"></a> 

## Fungsi Keanggotaan Waktu [↑](#up) <a class="anchor" id="Anggota_Waktu"></a>

### Subuh [↑](#up) <a class="anchor" id="Subuh"></a>

```python
def wktsbh(waktuin) :
    nilai_wktsbh = 0
    
    if (waktuin < 5 ): 
        nilai_wktsbh = 1
    elif (waktuin < 7): 
        nilai_wktsbh = (-((waktuin - 7) / (7 - 5)))
    else: 
        nilai_wktsbh = 0
    
    return nilai_wktsbh
```

### Pagi [↑](#up) <a class="anchor" id="Pagi"></a>

```python
def wktpg(waktuin):
    nilai_wktpg = 0
    
    if (5 < waktuin <= 7): 
        nilai_wktpg = ((waktuin - 5) / (7 - 5))
    elif (7 < waktuin <= 10):
        nilai_wktpg = 1
    elif (10 < waktuin <= 11): 
        nilai_wktpg = (-((waktuin - 11) / (11 - 10)))
    else: 
        nilai_wktpg = 0
        
    return nilai_wktpg
```

### Siang [↑](#up) <a class="anchor" id="Siang"></a>

```python
def wktsg(waktuin):
    nilai_wktsg = 0
    
    if (10 < waktuin <= 11): 
        nilai_wktsg = ((waktuin - 10) / (11 - 10))
    elif (11 < waktuin <= 14):
        nilai_wktsg = 1
    elif (14 < waktuin <= 15): 
        nilai_wktsg = (-((waktuin - 15) / (15 - 14)))
    else: 
        nilai_wktsg = 0
        
    return nilai_wktsg
```

### Sore [↑](#up) <a class="anchor" id="Sore"></a>

```python
def wktsr(waktuin):
    nilai_wktsr = 0
    
    if (14 < waktuin <= 15): 
        nilai_wktsr =  ((waktuin - 14) / (15 - 14))
    elif (15 < waktuin <= 17):
        nilai_wktsr =  1
    elif (17 < waktuin <= 18): 
        nilai_wktsr =  (-((waktuin - 18) / (18 - 17)))
    else: 
        nilai_wktsr =  0
        
    return nilai_wktsr
```

### Malam [↑](#up) <a class="anchor" id="Malam"></a>

```python
def wktmlm(waktuin):
    nilai_wktmlm = 0
    
    if (waktuin < 17): 
        nilai_wktmlm = 0
    elif (waktuin > 18): 
        nilai_wktmlm = 1
    else: 
        nilai_wktmlm = ((waktuin - 17) / (18 - 17))
        
    return nilai_wktmlm
```

___

## Fungsi Grafik Waktu [↑](#up) <a class="anchor" id="Grafik_Waktu"></a>

```python
def nilai_grafikwaktu(waktuin):

    print("Nilai Fuzzy Waktu")

    if wktsbh(waktuin) != 0:
        print("Subuh = " + str(wktsbh(waktuin)))
    if wktpg(waktuin) != 0:
        print("Pagi = " + str(wktpg(waktuin)))
    if wktsg(waktuin) != 0:
        print("Siang = " + str(wktsg(waktuin)))
    if wktsr(waktuin) != 0:
        print("Sore = " + str(wktsr(waktuin)))
    if wktmlm(waktuin) != 0:
        print("Malam = " + str(wktmlm(waktuin)))
```

```python
def grafikwaktu():
    print("Subuh < Subuh - Pagi < Pagi - Siang < Siang < Siang - Sore < Sore < Sore - Malam < Malam")
    wakturange = np.arange(0,24,1)
    plt.plot(wakturange, [wktsbh(x) for x in wakturange], color='Blue', label='Subuh')
    plt.plot(wakturange, [wktpg(x) for x in wakturange], color='Red', label='Pagi')
    plt.plot(wakturange, [wktsg(x) for x in wakturange], color='Green', label='Siang')
    plt.plot(wakturange, [wktsr(x) for x in wakturange], color='Magenta', label='Sore')
    plt.plot(wakturange, [wktmlm(x) for x in wakturange], color='Black', label='Malam')
    plt.title('Grafik Waktu')
    plt.xlabel('Waktu dalam Jam')
    plt.legend()
    plt.show()
```

```python
grafikwaktu()
```

# METHOD KELEMBABAN [↑](#up) <a class="anchor" id="Method_Kelembaban"></a>

## Fungsi Keanggotaan Kelembaban [↑](#up) <a class="anchor" id="Anggota_Kelembaban"></a>

### Kering [↑](#up) <a class="anchor" id="Kering"></a>

```python
def angkrg(kelembabanin):
    nilai_kering = 0
    
    if (kelembabanin < 35 ): 
        nilai_kering = 1
    elif (kelembabanin < 45): 
        nilai_kering = (-((kelembabanin - 45) / (45 - 35)))
    else: 
        nilai_kering = 0

    return nilai_kering
```

### Normal [↑](#up) <a class="anchor" id="Lembab"></a>

```python
def angnorm(kelembabanin):
    nilai_lembab = 0
    
    if (35 < kelembabanin <= 45): 
        nilai_lembab = ((kelembabanin - 35) / (45 - 35))
    elif (45 < kelembabanin <= 50):
        nilai_lembab = 1
    elif (50 < kelembabanin <= 65): 
        nilai_lembab = (-((kelembabanin - 65) / (65 - 50)))
    else: 
        nilai_lembab = 0

    return nilai_lembab
```

### Lembab [↑](#up) <a class="anchor" id="Basah"></a>

```python
def anglmb(kelembabanin):
    nilai_basah = 0
    
    if (kelembabanin < 50): 
        nilai_basah = 0
    elif (kelembabanin > 65): 
        nilai_basah = 1
    else: 
        nilai_basah = ((kelembabanin - 50) / (65 - 50))

    return nilai_basah
```

## Persamaan Kelembaban [↑](#up) <a class="anchor" id="Persamaan_Kelembaban"></a>

 <div class="alert alert-success">
<div>

$Kelembaban [Kering] = 
\left\{
    \begin{array}{ll}
       1 & \mbox{; }x \leq 35 \\
       {45-x \over 45-35} & \mbox{; } 35 < x < 45 \\
       0 & \mbox{; } x \geq 45
    \end{array}
\right.$ 

___

$Kelembaban [Normal] = 
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x \leq 35,OR,x \geq 65 \\
       {x-35 \over 45-35} & \mbox{; } 35 < x < 45 \\
       1 & \mbox{; }45  \leq x \leq 50\\
       {65-x \over 65-50} & \mbox{; } 50 < x < 65\\ 
    \end{array}
\right.$ 

___
    
$Kelembaban [Lembab] = 
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x \leq 50 \\
       {x-50 \over 65-50} & \mbox{; } 50 < x < 65 \\
       1& \mbox{; } x \geq 65
    \end{array}
\right.$

___

<div>

## Fungsi Grafik Kelembaban [↑](#up) <a class="anchor" id="Grafik_Kelembaban"></a>

```python
def nilai_grafikkelembaban(kelembabanin):
    print("\n")
    print("Nilai Fuzzy Kelembaban")

    if angkrg(kelembabanin) != 0:
        print("Kering = " + str(angkrg(kelembabanin)))
    if angnorm(kelembabanin) != 0:
        print("Normal = " + str(angnorm(kelembabanin)))
    if anglmb(kelembabanin) != 0:
        print("Lembab = " + str(anglmb(kelembabanin)))
```

```python
def grafikkelembaban():
    print("Kering < Kering - Normal < Normal - Lembab < Lembab")
    kelembabanrange = np.arange(0,100,1)
    plt.plot(kelembabanrange, [angkrg(x) for x in kelembabanrange], color='Red', label='Kering')
    plt.plot(kelembabanrange, [angnorm(x) for x in kelembabanrange], color='Green', label='Normal')
    plt.plot(kelembabanrange, [anglmb(x) for x in kelembabanrange], color='Blue', label='Lembab')
    plt.title('Grafik Kelembaban')
    plt.xlabel('Kelembaban Ruangan')
    plt.legend()
    plt.show()
```

# METHOD SUHU [↑](#up) <a class="anchor" id="Method_Suhu"></a>

## Fungsi Keangotaan Suhu [↑](#up) <a class="anchor" id="Anggota_Suhu"></a>

### Dingin [↑](#up) <a class="anchor" id="Dingin"></a>

```python
def shdingin(suhuin):
    nilai_dingin = 0
    
    if (suhuin < 12 ): 
        nilai_dingin =  1
    elif (suhuin < 19): 
        nilai_dingin = (-((suhuin - 19) / (19 - 12)))
    else: 
        nilai_dingin = 0
        
    return nilai_dingin
```

### Normal [↑](#up) <a class="anchor" id="Normal"></a>

```python
def shnormal(suhuin):
    nilai_normal = 0
    
    if (12 < suhuin <= 19): 
        nilai_normal = ((suhuin - 12) / (19 - 12))
    elif (19 < suhuin <= 20):
        nilai_normal = 1
    elif (20 < suhuin <= 27): 
        nilai_normal = (-((suhuin - 27) / (27 - 20)))
    else: 
        nilai_normal = 0
        
    return nilai_normal
```

### Hangat [↑](#up) <a class="anchor" id="Hangat"></a>

```python
def shhangat(suhuin):
    nilai_hangat = 0
    
    if (20 < suhuin <= 27): 
        nilai_hangat = ((suhuin - 20) / (27 - 20))
    elif (27 < suhuin <= 30):
        nilai_hangat = 1
    elif (30 < suhuin <= 36): 
        nilai_hangat = (-((suhuin - 36) / (36 - 30)))
    else: 
        nilai_hangat = 0
        
    return nilai_hangat
```

### Panas [↑](#up) <a class="anchor" id="Panas"></a>

```python
def shpanas(suhuin):
    nilai_panas = 0
    
    if (suhuin < 30): 
        nilai_panas = 0
    elif (suhuin > 36): 
        nilai_panas = 1
    else: 
        nilai_panas = ((suhuin - 30) / (36 - 30))
        
    return nilai_panas
```

## Fungsi Keanggotaan Kondisi [↑](#up) <a class="anchor" id="Fungsi_Kondisi"></a>

### Tutup [↑](#up) <a class="anchor" id="Tutup"></a>

```python
def kdstutup(sampel):
    if (sampel < 50 ): 
      return 1
    elif (sampel < 80): 
      return (-((sampel - 80) / (80 - 50)))
    else: 
      return 0
```

### Buka [↑](#up) <a class="anchor" id="Buka"></a>

```python
def kdsbuka(sampel):
    if (sampel < 50 ): 
      return 0
    elif (sampel > 80): 
      return 1
    else: 
      return (((sampel - 50) / (80 - 50)))
```

    \end{array}
\right.$
